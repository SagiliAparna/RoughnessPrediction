```python
def remove_outliers_iqr(df):
    """
    Remove outliers from a dataset using the Interquartile Range (IQR) method.

    Parameters:
        data (numpy array): The dataset from which outliers are to be removed.

    Returns:
        numpy array: Cleaned dataset with outliers removed.
        list: List of indices of identified outliers.
    """
    # Calculate the first quartile (Q1) and the third quartile (Q3) for each column
    Q1 = np.percentile(df,25, axis=0)
    Q3 = np.percentile(df, 75, axis=0)

    # Calculate the interquartile range (IQR) for each column
    IQR = Q3 - Q1

    # Define the lower and upper bounds for outliers for each column
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers for each column
    outliers_indices = np.where((df < lower_bound) | (df > upper_bound))

    # Remove outliers
    cleaned_data = df[~((df < lower_bound) | (df > upper_bound)).any(axis=1)]

    return  cleaned_data, outliers_indices


# Remove outliers using the function
cleaned_scores, outliers_indices = remove_outliers_iqr(df)

print("Indices of identified outliers:\n", outliers_indices) 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error

def perform_linear_regression(x):
    x_train, x_test, y_train, y_test = train_test_split(x, y1, test_size=0.2)
    model1 = LinearRegression()
    model1.fit(x_train, y_train)
    y_pred1 = model1.predict(x_test)
    mae = mean_absolute_error(y_test, y_pred1)
    rdtscore = r2_score(y_test, y_pred1)
    mse = mean_squared_error(y_test, y_pred1)
    return model1, mae, rdtscore, mse,y_pred1,y_test

def perform_decision_tree(x):
    x_train, x_test, y_train, y_test = train_test_split(x, y1, test_size=0.2)
    model2 = DecisionTreeRegressor()
    model2.fit(x_train, y_train)
    y_pred2 = model2.predict(x_test)
    mae = mean_absolute_error(y_test, y_pred2)
    rdtscore = r2_score(y_test, y_pred2)
    mse = mean_squared_error(y_test, y_pred2)
    return model2, mae, rdtscore, mse,y_pred2

def perform_random_forest( x):
    x_train, x_test, y_train, y_test = train_test_split(x, y1, test_size=0.2)
    model3 = RandomForestRegressor()
    model3.fit(x_train, y_train)
    y_pred3 = model3.predict(x_test)
    mae = mean_absolute_error(y_test, y_pred3)
    rdtscore = r2_score(y_test, y_pred3)
    mse = mean_squared_error(y_test, y_pred3)
    return model3, mae, rdtscore, mse

def perform_k_neighbors_regression(x):
    x_train, x_test, y_train, y_test = train_test_split(x, y1, test_size=0.2)
    model4 = KNeighborsRegressor()
    model4.fit(x_train, y_train)
    y_pred4 = model4.predict(x_test)
    mae = mean_absolute_error(y_test, y_pred4)
    rdtscore = r2_score(y_test, y_pred4)
    mse = mean_squared_error(y_test, y_pred4)
    return model4, mae, rdtscore, mse

# Assuming you have defined  x and   elsewhere
lr_model1, mae_lr, r2_lr, mse_lr,y_pred1,y_test = perform_linear_regression(x)
dt_model2, mae_dt, r2_dt, mse_dt,y_pred2 = perform_decision_tree( x)
rf_model3, mae_rf, r2_rf, mse_rf = perform_random_forest( x)
knr_model4, mae_knr, r2_knr, mse_knr = perform_k_neighbors_regression( x)

print("Linear Regression:")
print("Mean Absolute Error:", mae_lr)
print("R^2 Score:", r2_lr)
print("Mean Square Error:", mse_lr)
print("Root Mean Square Error:", mse_lr**(1/2))
print(y_pred1)

print("\nDecision Tree Regression:")
print("Mean Absolute Error:", mae_dt)
print("R^2 Score:", r2_dt)
print("Mean Square Error:", mse_dt)
print("Root Mean Square Error:", mse_dt**(1/2))


print("\nRandom Forest Regression:")
print("Mean Absolute Error:", mae_rf)
print("R^2 Score:", r2_rf)
print("Mean Square Error:", mse_rf)
print("Root Mean Square Error:", mse_rf**(1/2))


print("\nK Neighbors Regression:")
print("Mean Absolute Error:", mae_knr)
print("R^2 Score:", r2_knr)
print("Mean Square Error:", mse_knr)
print("Root Mean Square Error:", mse_knr**(1/2))
```
